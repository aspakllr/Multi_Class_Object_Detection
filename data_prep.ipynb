```python
import scipy.io
import pandas as pd
import sys
import os
from os.path import join as jp
import numpy as np
import csv
import cv2
import matplotlib.pyplot as plt
```

#### Create csv files containing all the necessary information about each object
* The info was derived by the original mat files provided in the Caltech 101 dataset (http://www.vision.caltech.edu/Image_Datasets/Caltech101/).
* The process produces one csv file for each object category, that contains the info (names, bbox coordinates, class) about the images.

```python
data_path = "/content/drive/MyDrive/Colab Notebooks/Multi_Class_Object_Detection/annotations" # Path to mat files
csv_path = "/content/drive/MyDrive/Colab Notebooks/Multi_Class_Object_Detection/annotations2"


def create_csv(data_path, csv_path, category):

  dataset = []
  
  for f in (os.listdir(data_path)):

    # Load annotations file
    mat = scipy.io.loadmat(jp(data_path, f))

    # Get image name
    name = "image" + f[10:-4] + ".jpg"
    # print("File", name)

    # Get bbox coordinates
    bbox = mat['box_coord'] # shape (1,4)
    bbox = np.squeeze(bbox, axis=0) # remove axis
           
    # Create numpy array containing info: [name, x1, y1, x2, y2, class] 
    data = [name, bbox[2], bbox[0], bbox[3], bbox[1], category]
    dataset.append(data)
    # print(data, "\n")

  # Save data in csv
  with open(csv_path, 'w') as f:
    writer = csv.writer(f)
    for row in dataset:
      writer.writerow(row)
  f.close()  

create_csv(jp(data_path,"annotations/octopus"), jp(data_path,"octopus.csv"), "octopus")
create_csv(jp(data_path,"annotations/butterfly"), jp(data_path,"butterfly.csv"), "butterfly")
create_csv(jp(data_path,"annotations/hedgehog"), jp(data_path,"hedgehog.csv"), "hedgehog")
```
